## Introduction and Model Selection

For this analysis I have chosen to model a spring-mass system. Specifically, I will model the vertical motion of a car after hitting a speed bump using a quarter-car suspension model consisting of one wheel and the associated mass. This is standard in mechanical and automotive engineering. A typical quarter-car mass is approximately 250–350 kg based on standard passenger vehicle weights (Rajamani, 2012). Suspension spring constants are commonly in the range of 10,000–20,000 N/m, and damping coefficients around 1,000–2,000 N·s/m (Gillespie, 1992). For this analysis, I will take the mass m to be 300 kg, the spring constant k to be 15,000 N/m, and the damping coefficient c to be 1200 N·s/m. Mass is self explanatory. The spring constant is how stiff a spring is, which is measured by the amount of force to compress it a certain distance. The damping term (the damping coefficient times velocity in the ODE) represents energy dissipation due to frictional and fluid effects in the shock absorber.

The governing second-order ordinary differential equation for a general spring-mass system is of the form $$m \frac{d^{2} y}{d t^{2}}+c \frac{d y}{d t}+k y=0$$ When plugging in for this specific scenario, we get the equation $$300 \frac{d^{2} y}{d t^{2}}+1200 \frac{d y}{d t}+15000 y=0$$ Now, if we let $$y_1 = y, \quad y_2 = \frac{dy}{dt}$$
