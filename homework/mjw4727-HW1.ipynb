importing relevant libraries

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

monte carlo pi estimate function

```python
def monteCarlo (numSamples):
    x = 0
    y = 0
    counter = 0
    for s in range (0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(x*x + y*y <= 1):
            counter = counter + 1
    return 4*(counter/numSamples)
```

iterating from 10 to 10000; making error list

```python
x = np.arange(10, 10000, 20)
y = np.zeros(len(x))
for i in range(0,len(x)):
    y[i] = monteCarlo(x[i])
error = abs(y-math.pi)
```

plotting (added log plot to make the trend easier to see)

```python
plt.subplot(1,2,1)
plt.scatter(x,error, 2)
plt.title("Error vs Sample Size")
plt.subplot(1,2,2)
plt.scatter(x,np.log(error), 2)
plt.title("log(Error) vs Sample Size")
plt.show
```
