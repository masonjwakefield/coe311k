libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

sin taylor series centered at zero function

```python
def sinTaylor(x, numberOfTerms):
    ret = 0
    for n in range (0,numberOfTerms):
        ret = ret + ((-1)**n)*(x**(2*n+1))/math.factorial(2*n+1)
    return ret
```

x=0 to x=10 by steps of .1

```python
values = np.arange(0,10.1,.1)
```

run values through np.sin and taylor series with 1,2,3,4 terms

```python
exact = np.sin(values)
oneTerm = sinTaylor(values, 1)
twoTerm = sinTaylor(values, 2)
threeTerm = sinTaylor(values, 3)
fourTerm = sinTaylor(values, 4)
```

find absolute errors

```python
oneTermError = abs(oneTerm-exact)
twoTermError = abs(twoTerm-exact)
threeTermError = abs(threeTerm-exact)
fourTermError = abs(fourTerm-exact)
```

plotting results

```python
plt.plot(values, oneTermError, "red")
plt.plot(values, twoTermError, "orange")
plt.plot(values, threeTermError, "green")
plt.plot(values, fourTermError, "blue")
plt.yscale("log")
plt.ylabel("Absolute Error")
plt.xlabel("x-value")
plt.legend("1234")
```

No matter the number of terms you use, your absolute error will increase as you move away from the center of your taylor series (in this case x=0). The absolute error increases with higher x-values at an approximately exponential rate. 
With lower x-values, using more terms produces much smaller absolute errors. However, above about x=5, using more terms produces much larger absolute errors. This trend continues to higher and higher x-values, with more terms producing more error.
