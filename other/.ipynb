```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)
```

```python
plt.plot(x, np.sin(x), "b-", x, 2*np.sin(x), "m-")
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, np.sin(x), 'b')
plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(1, 3, 3)
plt.plot(x, np.tan(x), 'r')
plt.show
```

```python
plt.scatter(x, np.sin(x))
plt.show
```

```python
def f1(x,y):
    return (x*x+y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(45, 0)
```
