libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Part 1

copied code

```python
# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

plotting

```python
plt.scatter(X,Y,20,"r")
plt.plot(X,m*X+c)
plt.title("Least Squares Regression Line")
plt.show()
```

# Part 2

now with random vectors

```python
rng = np.random.default_rng()

# Data points
X = rng.random(5)
X = 5*X
X.sort()
Y = rng.random(5)
Y = 5*Y

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

plotting

```python
plt.scatter(X,Y,20,"r")
plt.plot(X,m*X+c)
plt.title("Least Squares Regression Line")
plt.show()
```

# Part 3

randomly generating new line, logging m & c

```python
truem = 5 * rng.random()
truec = 5 * rng.random()

print(f"true m: {truem}")
print(f"true c: {truec}")

X = np.arange(0,1,.01)
rand = rng.random(100)
rand = 2*(rand-0.5)
Y = truem*X+truec+rand
```

calculating using previous code

```python
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

plotting

```python
plt.scatter(X,Y,5,"r")
plt.plot(X,m*X+c)
plt.title("Least Squares Regression Line")
plt.show()
```

printing residuals

```python
resid = Y-(m*X+c)
print(resid)
```

printing max and min residuals

```python
print(f"Max residual: {resid.max()}")
print(f"Min residual: {resid.min()}")
```

The least squares regression was fairly accurate to the true values of m and c. However, from testing it is less accurate with lower values of m and c as the resulting randomized points are a blob about the x-axis with no definite shape.
