libraries

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

building matrices

```python
a1 = np.random.rand(3,3)
a2 = np.random.rand(3,3)

b1 = np.random.rand(6,6)
b2 = np.random.rand(6,6)

c1 = np.random.rand(12,12)
c2 = np.random.rand(12,12)

d1 = np.random.rand(24,24)
d2 = np.random.rand(24,24)

e1 = np.random.rand(48,48)
e2 = np.random.rand(48,48)

f1 = np.random.rand(96,96)
f2 = np.random.rand(96,96)

g1 = np.random.rand(192,192)
g2 = np.random.rand(192,192)

h1 = np.random.rand(384,384)
h2 = np.random.rand(384,384)
```

naive matrix mult algorithm

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

timing naive algorithm

```python
trial = 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(a1,a1)
endtime = time.perf_counter()
runtimea = endtime-starttime
print(f"Trial {trial} runtime: {runtimea}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(b1,b1)
endtime = time.perf_counter()
runtimeb = endtime-starttime
print(f"Trial {trial} runtime: {runtimeb}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(c1,c1)
endtime = time.perf_counter()
runtimec = endtime-starttime
print(f"Trial {trial} runtime: {runtimec}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(d1,d1)
endtime = time.perf_counter()
runtimed = endtime-starttime
print(f"Trial {trial} runtime: {runtimed}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(e1,e1)
endtime = time.perf_counter()
runtimee = endtime-starttime
print(f"Trial {trial} runtime: {runtimee}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(f1,f1)
endtime = time.perf_counter()
runtimef = endtime-starttime
print(f"Trial {trial} runtime: {runtimef}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(g1,g1)
endtime = time.perf_counter()
runtimeg = endtime-starttime
print(f"Trial {trial} runtime: {runtimeg}")
trial = trial + 1

starttime = time.perf_counter()
answer = naive_matrix_multiplication(h1,h1)
endtime = time.perf_counter()
runtimeh = endtime-starttime
print(f"Trial {trial} runtime: {runtimeh}")
trial = trial + 1
```

plotting both

```python
ynaive = [runtimea,runtimeb,runtimec,runtimed,runtimee,runtimef,runtimeg,runtimeh]
ynumpy = [runtimea2,runtimeb2,runtimec2,runtimed2,runtimee2,runtimef2,runtimeg2,runtimeh2]
x = np.arange(1,9,1)

plt.semilogy(x,ynaive,label = 'Naive Algorithm')
plt.semilogy(x,ynumpy, label = 'Numpy Algorithm')
plt.legend()
plt.show()
```

from observing the plot, it becomes very clear that the naive algorithm takes much much longer with larger matrices than the optimized numpy algorithm. this is because it has O(n^3) complexity, which is almost certainly worse than the efficiency of the numpy algorithm which scales much better
