import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

defining function (dy/dx = -2y)

```python
# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y
```

from my experience in differential equations, i found the original equation to be y=e^(-2x), this is used to find errors below

iterate through Euler's Method using different h values

```python
for j in [.1,.5,1,1.5]:
    # Define parameters
    h = j  # step size
    x_start, x_end = 0, 10
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    # Plot the results
    plt.plot(x_values, y_values, label="Euler's Method Approximation")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    
    # Error tracking
    point_error = np.abs(y_values - np.e**(-2*x_values))
    global_error = np.sum(point_error)
    accumulated_error = 0
    cumulative_error = []
    for num in point_error:
        accumulated_error += num
        cumulative_error.append(accumulated_error)
    
    # Plot the error
    plt.plot(x_values,point_error, 'red', label = 'Point Error (abs)')
    plt.plot(x_values,cumulative_error, 'black', label = 'Cumulative Global Error (sum of abs)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Error of Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
    
    # Log useful information
    print(f"Step size: {h}")
    print(f"Error at each point: {point_error}")
    print(f"Global error: {global_error}")
```
