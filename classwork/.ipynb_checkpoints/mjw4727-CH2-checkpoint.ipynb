```python
def convertToBinary(x):
    ogx = x
    equalZero = False
    backwards = []
    one = "1"
    zero = "0"
    ret = ""
    
    while(equalZero == False):
        backwards.append(x%2)
        if(x%2 == 0):
            x = x/2
        else:
            x = x-1
            x = x/2 
        if(x == 0):
            equalZero = True
    
    for i in backwards:
        if(i == 1):
            ret = one + ret
        if(i == 0):
            ret = zero + ret
            
    print(f"{ogx} in binary is {ret}")
```

```python
convertToBinary(13)
```
