diagonal dominance algorithm

```python
def diagdom(A):
    dom = True
    absmat = np.abs(A)
    
    for i in range(0, len(A)):
        if(absmat[i,i] < (np.sum(absmat[i,:]) - absmat[i,i])):
            dom = False
            
    print(A)
    if(dom == True):
        print("Is the matrix diagonally dominant? True")
    else:
        print("Is the matrix diagonally dominant? False")
    return(dom)
```

```python
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
diagdom(A)
```
