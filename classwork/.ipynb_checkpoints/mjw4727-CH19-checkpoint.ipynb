import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

defining function (dy/dx = -100y + 3000 - 2000e^-t)

```python
# Function representing dy/dt = -1000y + 3000 - 2000e^-t
def f(t, y):
    return (-1000 * y) + (3000) - (2000 * (np.exp(-t)))
```

iterate through Euler's Method using different h values, also plot analytical solution

```python
for j in [.1,.01,.001, .0001, .00001, .000001]:
    # Define parameters
    h = j  # step size
    t_start, t_end = 0, 30
    num_steps = int((t_end - t_start) / h)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 0  # y(0) = 0

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])
        
    # Plot the results
    plt.plot(t_values, y_values, label=f"Forward Euler with h={h}")
    plt.xlabel("t")
    plt.ylabel("y")
    plt.title(f"Forward Euler with h={h} for Stiff Equation")
    plt.grid(True)
    plt.show()
    
# Plot the analytical solution (obtained from slides)
plt.plot(t_values, 3 - (.998*(np.e**(-1000*t_values))) - (2.002 * (np.e**(-t_values))), label = "Analytical Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Analytical Solution of Stiff Equation")
plt.grid(True)
plt.show()
```

Note: The forward Euler method blows up with h=.1 and h=.01. With h=.001 and smaller, the forward Euler plots and the analytical solution are almost exactly the same. They have been plotted on separate graphs because they are so similar as to be indistinguishable on the same plot. Using h values lower than .000001 results in runtimes that are too high and not worth running due to their similarity to prior h values.
