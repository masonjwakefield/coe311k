importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

run demo code with different h values

```python
for h in [.01, .02, .03, .05, .1, .2, .3, .5]:
    # Define the parameters
    k = 1.0        # Constant in the equation (e.g., spring constant)
    x_start = 0    # Start of the interval
    x_end = 10     # End of the interval
    y_0 = 1.0      # Initial condition for y
    v_0 = 0.0      # Initial condition for dy/dx
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot the results
    plt.plot(x_values, y_values, label=f'Position y(x) with h={h}')
    plt.plot(x_values, v_values, label=f'Velocity v(x) with h={h}', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method (varying values of h)')
plt.grid(True)
plt.legend(fontsize = 4.5, loc = 'lower left')
plt.figure(figsize=(10, 5))
plt.show()
```

The figures seem to become unstable around a value of h=.5, showing the first true divergence from the rest at h=.5. The model is becoming more unstable with higher h values, most likely being completely useless with h values any higher than 0.5.

run demo code with varying values of k

```python
for k in [1, 2, 3, 4]:
    # Define the parameters
    h = 0.01       # Step size
    x_start = 0    # Start of the interval
    x_end = 10     # End of the interval
    y_0 = 1.0      # Initial condition for y
    v_0 = 0.0      # Initial condition for dy/dx
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot the results
    plt.plot(x_values, y_values, label=f'Position y(x) with k={k}')
    plt.plot(x_values, v_values, label=f'Velocity v(x) with k={k}', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method (varying values of k)')
plt.grid(True)
plt.legend(fontsize = 4.5, loc = 'lower left')
plt.figure(figsize=(10, 5))
plt.show()
```

When playing with the value of k, changes can be seen in the maximum velocity reached and the period of the oscillation. With increasing values of k, the maximum velocity of each oscillation increases and the period (time for each oscillation) decreases. The amplitude oscillations does not change.

run demo code with k=10 and h=.01

```python
# Define the parameters
k = 10        # Constant in the equation (e.g., spring constant)
h = 0.01       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx
# Create arrays to store x, y, and v values
x_values = np.arange(x_start, x_end + h, h)
y_values = np.zeros(len(x_values))
v_values = np.zeros(len(x_values))
# Set initial conditions
y_values[0] = y_0
v_values[0] = v_0

# Euler's Method loop
for i in range(1, len(x_values)):
    # Current values
    x_n = x_values[i - 1]
    y_n = y_values[i - 1]
    v_n = v_values[i - 1]    
    # Update equations
    y_values[i] = y_n + h * v_n
    v_values[i] = v_n + h * (-k * y_n)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='Position y(x)')
plt.plot(x_values, v_values, label='Velocity v(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method (k=10, h=.01)')
plt.legend()
plt.grid(True)
plt.show()
```

I do not believe that higher h values would be able to solve this second order ODE with k=10. With h=.01, you can already see error forming on the right side of the graph (x>5) because the amplitude of the oscillations is increasing. The amplitude of the oscillations should remain constant, so error exists with an h as low as .01. Any higher may be wildly inaccurate.
