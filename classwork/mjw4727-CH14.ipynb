libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

define the function and the domain

```python
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points

x = np.linspace(a, b, N)
```

approximations

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
dfdxcentral = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
dfdxforward = (f(x + h) - f(x))/h

# Backward Difference Approximation
dfdxbackward = (f(x) - f(x - h))/h
```

plot the approximation and the true derivative

```python
plt.plot(x,dfdxcentral, label = 'Central')
plt.plot(x,dfdxforward, label = 'Forward')
plt.plot(x,dfdxbackward, label = 'Backward')
plt.plot(x,np.cos(x), label = 'True (cos(x))')
plt.title('Approxations of sin(x) Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

note: dfdxcentral is so similar to the true dfdx=cos(x) that it is not easily visible on the graph
