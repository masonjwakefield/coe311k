importing libraries

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

monte carlo function

```python
def monteCarlo (numSamples):
    x = 0
    y = 0
    counter = 0
    for s in range (0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(x*x + y*y <= 1):
            counter = counter + 1
    return 4*(counter/numSamples)
```

making lists; finding errors

```python
estpi = 0
abserror = []
relerror = []
for i in range (0,100):
    estpi = monteCarlo(1000)
    abserror.append(abs(estpi-math.pi))
    relerror.append(abs((estpi-math.pi)/math.pi))
```

plotting absolute and relative error

```python
plt.subplot(1,2,1)
plt.hist(abserror, 10)
plt.title("Absolute Error Distribution")
plt.subplot(1,2,2)
plt.hist(relerror, 10)
plt.title("Relative Error Distribution")
plt.show
```
